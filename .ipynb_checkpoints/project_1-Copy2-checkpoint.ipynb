##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
from math import pow

def calc_var(data):
    len_data = len(data)
    avg_rate = sum([int(a_list[1]) for a_list in data])/len_data
    avg_verbal = sum([int(a_list[2]) for a_list in data])/len_data
    avg_math = sum([int(a_list[3]) for a_list in data])/len_data
    
    var_rate = sum([(a_list[1] - pow(avg_rate, 2)) for a_list in data]) / len_data
    var_verbal = sum([(a_list[2] - pow(avg_verbal, 2)) for a_list in data]) / len_data
    var_math = sum([(a_list[3] - pow(avg_math, 2)) for a_list in data]) / len_data
    
    print "var of rate = " + str(var_rate)
    print "var_of_verbal = " + str(var_verbal)
    print "var_of_math = " + str(var_math)

calc_var(data)
```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = math_scores
y = verbal_scores

plt.ylabel('Verbal_Scores')
plt.xlabel('Math_scores')
plt.title('SAT scores in the US')

plt.scatter(x, y, alpha=0.5)
coefs = np.polyfit(x, y, deg=1)
fit_y = np.lib.polyval(coefs, x)
plt.plot(x, fit_y, 'r--')
plt.show()
```

##### 21. Create box plots for each variable. 

```python
data_to_plot=[verbal_scores,math_scores, rates]
fig = plt.figure(1, figsize=(15, 10))
ax = fig.add_subplot(111)
bp = ax.boxplot(data_to_plot, patch_artist=True)

for box in bp['boxes']:
    box.set(color='#CCB974', linewidth=2)
    box.set( facecolor = '#7A68A6')
for whisker in bp['whiskers']:
    whisker.set(color='#4C72B0', linewidth=2)
for cap in bp['caps']:
    cap.set(color= '#8C0900', linewidth=2)
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=1)
    
ax.set_xticklabels(['verbal_scores', 'math_scores', 'rates'])

fig.savefig('fig1.png', bbox_inches='tight')

```
